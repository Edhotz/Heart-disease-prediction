## Vector Operations in Python


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Plotting a vector

plt.quiver(0, 0, 4, 5)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
plt.quiver(0, 0, 4, 5, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, -3, -6, angles='xy', scale_units='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

### Adding of two vectors


```python
vector1 = np.array([4, 5])
vector2 = np.array([3, -6])

vector_sum = vector1 + vector2
vector_sum
```

```python
plt.quiver(0, 0, 4, 5, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, 3, 6, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, 5, 3, angles='xy', scale_units='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
vector1 = np.array([4, 5])
vector2 = np.array([3, -6])

vector_diff = vector1 - vector2
vector_diff
```

```python
plt.quiver(0, 0, 2, 1, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, 3, -2, angles='xy', scale_units='xy', scale=1, color='y')
plt.quiver(0, 0, 1, 5, angles='xy', scale_units='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
## Multipliying vector by a scalar
vector1 = np.array([4, 5])

vector2 = 2 * vector1
vector2
```

```python
plt.quiver(0, 0, 1, 3, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, 4, 6, angles='xy', scale_units='xy', scale=1, color='y')
plt.quiver(0, 0, 8, 5, angles='xy', scale_units='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
vector1 = np.array([4, 5])


vector2 = 0.5 * vector1
vector2
```

```python
plt.quiver(0, 0, 2, 2.5, angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, 4, 1, angles='xy', scale_units='xy', scale=1, color='y')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```
